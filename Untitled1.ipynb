```python
PDF_PATH_1 = "/content/Operation instruction_sample 1.pdf"
PDF_PATH_2 = "/content/Operation instruction_sample 2.pdf"
```

```python
from unstructured.partition.pdf import partition_pdf

# Process first PDF
elements_1 = partition_pdf(
    filename=PDF_PATH_1,
    strategy="hi_res",
    ocr_languages="eng",
    infer_table_structure=False,
)
raw_text_1 = "\n".join(e.text for e in elements_1 if e.text and e.text.strip())
print(f"Extracted {len(raw_text_1):,} characters from {PDF_PATH_1}")
```

```python
# Process second PDF
elements_2 = partition_pdf(
    filename=PDF_PATH_2,
    strategy="hi_res",
    ocr_languages="eng",
    infer_table_structure=False,
)
raw_text_2 = "\n".join(e.text for e in elements_2 if e.text and e.text.strip())
print(f"Extracted {len(raw_text_2):,} characters from {PDF_PATH_2}")

# Combine raw text from both PDFs
raw_text = raw_text_1 + "\n" + raw_text_2
print(f"Total extracted {len(raw_text):,} characters from both PDFs")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
import uuid, rich
from nltk.tokenize import sent_tokenize
from typing import Optional, List
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0)

class AgenticChunker:
    def __init__(self, llm, id_len=5):
        self.llm, self.id_len = llm, id_len
        self.chunks = {}

    _route_prompt = ChatPromptTemplate.from_messages([
        ("system","Return ONLY the id of the chunk the proposition fits, or 'NONE'."),
        ("user","{outline}"),
        ("user","Proposition: {prop}")
    ])
    _sum_prompt   = ChatPromptTemplate.from_messages([("system","One‑sentence summary:"),("user","{props}")])
    _title_prompt = ChatPromptTemplate.from_messages([("system","2‑5 word title:"),("user","{summary}")])

    def _find_chunk(self, prop):
        if not self.chunks: return None
        outline = "\n".join(f"[{cid}] {c['title']}: {c['summary']}"
                             for cid,c in self.chunks.items())
        resp = (self._route_prompt|self.llm).invoke({"outline":outline,"prop":prop}).content.strip()
        return resp if resp in self.chunks else None

    def _new_chunk(self, prop):
        cid = str(uuid.uuid4())[:self.id_len]
        summary = (self._sum_prompt|self.llm).invoke({"props":prop}).content.strip()
        title   = (self._title_prompt|self.llm).invoke({"summary":summary}).content.strip()
        self.chunks[cid] = {"title":title,"summary":summary,"props":[prop]}

    def add(self, propositions:List[str]):
        for p in propositions:
            cid = self._find_chunk(p)
            if cid: self.chunks[cid]["props"].append(p)
            else:   self._new_chunk(p)

    def strings(self):  # list of chunk texts
        return [" ".join(c["props"]) for c in self.chunks.values()]

    def show(self):
        for cid,c in self.chunks.items():
            rich.print(f"[bold cyan]{cid}[/] — {c['title']}  • {c['summary']}")
            for p in c["props"]: rich.print("   •",p)
            print()

# sentence → propositions
propositions = sent_tokenize(raw_text)
ac = AgenticChunker(llm)
ac.add(propositions)      # use all sentences; slice if enormous
ac.show()
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain.docstore.document import Document

docs = [Document(page_content=s, metadata={"src":"pdf"}) for s in ac.strings()]
emb  = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
vectordb = Chroma.from_documents(docs, emb, collection_name="pdf_chunks")

# Custom retriever with cosine similarity threshold
class CosineSimilarityRetriever:
    def __init__(self, vectordb, embedder, threshold=0.01, k=10): # k is max documents to retrieve before filtering
        self.vectordb = vectordb
        self.embedder = embedder
        self.threshold = threshold
        self.k = k

    def get_relevant_documents(self, query: str) -> List[Document]:
        query_embedding = self.embedder.embed_query(query)
        results = self.vectordb.similarity_search_with_score(query, k=self.k)

        filtered_docs = []
        for doc, score in results:
            # ChromaDB's similarity_search_with_score typically returns L2 distance.
            # We need to convert it to cosine similarity. A smaller L2 distance
            # means higher similarity.
            # For simplicity, we are assuming a direct relationship or will
            # recalculate cosine similarity if needed.
            # However, for thresholding, a direct score comparison is often sufficient
            # if we know the range of the scores from ChromaDB.
            # To be precise with cosine similarity, we would re-embed and calculate.
            # Let's assume for now that a score < threshold is what we want to keep
            # if the scores are L2 distances, meaning smaller is better.

            # If ChromaDB returns true cosine similarity, then score > threshold.
            # If ChromaDB returns L2 distance, then score < threshold.

            # For a general approach, let's re-embed and calculate cosine similarity
            # This is more robust for a "cosine similarity threshold".

            doc_embedding = self.embedder.embed_query(doc.page_content)
            cos_sim = cosine_similarity([query_embedding], [doc_embedding])[0][0]

            if cos_sim >= self.threshold:
                filtered_docs.append(doc)
        return filtered_docs

retriever = CosineSimilarityRetriever(vectordb, emb, threshold=0.01, k=10) # Set your desired threshold and k
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough, RunnableLambda
from langchain_core.output_parsers import StrOutputParser
from IPython.display import Markdown

prompt = ChatPromptTemplate.from_template("""
You are an AI assistant specialized in answering questions based *strictly* on the provided context.
Answer the user's question accurately and concisely using *only* the information present in the context.
If relevant information is spread across different parts of the context, synthesize it into a single coherent answer.
If the answer cannot be found within the provided context, explicitly state: "I’m not sure based on the current information."
Do not make up information.

Respond in a way that is easy for beginners to understand:
- Understand the user's question carefully.
- Use simple, clear language so even a complete beginner can understand.
- Break down complex ideas step by step.
- Be friendly and talk like a real human — like you're talking to a curious person who needs help.
- If you need more information from the user to give a better answer, politely ask them for it.
- **Never output numbers (like spare part numbers) without explaining what they mean and why they're used.**


- Understand the user's problem clearly.
- Provide accurate, solution-oriented responses.
- Use a step-by-step format if it helps the user follow the solution better.

Context:
{context}

Question: {question}
""")

qa_chain = (
    {"context": RunnableLambda(retriever.get_relevant_documents), "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)


from IPython.display import Markdown, display

while True:
    q = input("\n❓ Ask (or 'exit'): ")
    if q.lower() == "exit":
        break
    answer = qa_chain.invoke(q)
    display(Markdown(f"**Answer:** {answer}"))
```
